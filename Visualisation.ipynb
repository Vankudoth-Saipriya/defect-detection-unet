```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
def visualize_images_cv2(title, original_dir, gt_dir, pred_dir, layers):
    n = len(layers)
    fig, axes = plt.subplots(n, 3, figsize=(12, 4 * n))
    fig.suptitle(title, fontsize=20)

    for i, layer_id in enumerate(layers):
        layer_name = f"layer_{layer_id}.png"

        # Construct full paths
        orig_path = os.path.join(original_dir, layer_name)
        gt_path = os.path.join(gt_dir, layer_name)
        pred_path = os.path.join(pred_dir, layer_name)

        # Load using OpenCV (BGR), convert to RGB for matplotlib
        orig_img = cv2.imread(orig_path)
        gt_img = cv2.imread(gt_path)
        pred_img = cv2.imread(pred_path)

        # Convert to RGB
        orig_img = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
        gt_img = cv2.cvtColor(gt_img, cv2.COLOR_BGR2RGB)
        pred_img = cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB)

        # Plot
        ax_row = axes[i] if n > 1 else axes
        ax_row[0].imshow(orig_img)
        ax_row[0].set_title(f"Original - {layer_name}")

        ax_row[1].imshow(gt_img)
        ax_row[1].set_title("Ground Truth")

        ax_row[2].imshow(pred_img)
        ax_row[2].set_title("Prediction")

        for ax in ax_row:
            ax.axis("off")

    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()

```

```python
# ----- Powder Bed Defect Detection -----
powder_bed_layers = [1656, 1664, 1669, 1682, 1684]
visualize_images_cv2(
    title="Powder Bed Defect Detection",
    original_dir=r"/content/drive/MyDrive/Info_Project/Defect_Detection/DataSets/Data.Splitting/Powder_Bed_Defect_Detection/test/Img.After.Powder.Spread",
    gt_dir=r"/content/drive/MyDrive/Info_Project/Defect_Detection/DataSets/Data.Splitting/Powder_Bed_Defect_Detection/ground_truths",
    pred_dir=r"/content/drive/MyDrive/Info_Project/Defect_Detection/DataSets/Data.Splitting/Powder_Bed_Defect_Detection/predicted",
    layers=powder_bed_layers
)
```

```python
# ----- After Melting Defect Detection -----
after_melting_layers = [54, 56, 74, 77, 83]
visualize_images_cv2(
    title="After Melting Defect Detection",
    original_dir=r"/content/drive/MyDrive/Info_Project/Defect_Detection/DataSets/Data.Splitting/After_Melting_Defect_Detection/test/Img.After.Melting",
    gt_dir=r"/content/drive/MyDrive/Info_Project/Defect_Detection/DataSets/Data.Splitting/After_Melting_Defect_Detection/ground_truths",
    pred_dir=r"/content/drive/MyDrive/Info_Project/Defect_Detection/DataSets/Data.Splitting/After_Melting_Defect_Detection/predicted",
    layers=after_melting_layers
)
```
